```python
df["review"] = df["review"].str.replace("  ", "")

df
```

```python
# 공백 기준 분할

df2 = df["review"]

df2
```

```python
# 글자 2개 이상인 경우만 출력

df_list = [] 
for word in df2:
    if len(word) > 2:
        df_list.append(word)

print(df_list)
```

```python
# 리스트 다시 데이터프레임화

df_list2 = pd.DataFrame(zip(df_list))

display(df_list2)
```
